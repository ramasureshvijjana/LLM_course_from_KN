# Spitting The Documents : RAG STAGE - 2

```python
# Loading text data

from langchain_community.document_loaders import TextLoader

# Loading the text doc.
text_doc = TextLoader(r"..\01_data_ingestion\speech.txt").load()
print(text_doc)
```

```python
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(separator="\n\n", chunk_size = 100, chunk_overlap=10)

splitted_docs = splitter.split_documents(text_doc)

print(splitted_docs)
```
