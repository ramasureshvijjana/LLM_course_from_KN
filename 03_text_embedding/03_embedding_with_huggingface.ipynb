# Text Embedding with HuggingFace

```python
import os
from dotenv import load_dotenv

load_dotenv()
os.environ['HF_TOKEN'] = os.getenv("HF_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

hf_embedder = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
# Making input documents.
embedded_docs = hf_embedder.embed_documents(["This is Rama and I am a Data scientist",
"This is Naveen and I am a CA"])

print(embedded_docs[0])
```

```python
# Making input documents.
f_embd = hf_embedder.embed_query("Data scientist is job having by Rama")

print(f_embd)
```
