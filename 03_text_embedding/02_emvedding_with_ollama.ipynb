# Text Embedding with Ollama Model

```python
from langchain_community.embeddings import OllamaEmbeddings

# Initiating the OllamaEmbeddings model.
ollama_embedder = (OllamaEmbeddings(model = 'llama3.2:1b'))
print(ollama_embedder)
```

```python
# Making input documents.
embedded_docs = ollama_embedder.embed_documents(["This is Rama and I am a Data scientist",
"This is Naveen and I am a CA"])

print(embedded_docs[0])
```

```python
# Making input documents.
f_embd = ollama_embedder.embed_query("Data scientist is job having by Rama")

print(f_embd)
```
